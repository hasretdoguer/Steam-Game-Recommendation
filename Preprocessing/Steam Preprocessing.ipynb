## Data Cleaning
#### Extract total_review and positive percentage data from rating column

```python
df.dropna(axis = 0,inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
import re
count = 0
total_review = []
positive_percentage = []
for i in df.rating:
    string = str(i)
    string = string.replace(",","")
    en_str = re.findall(r"\buser\b",string)
    if en_str:
        try:
            total_review.append(re.findall(r'[^%](\d+)',string)[1])
        except:
            total_review.append("0")
        try:
            positive_percentage.append(re.findall(r'(\d+)',string)[0])
        except:
            positive_percentage.append("0")
    else:
        try:
            total_review.append(re.findall(r'[^%](\d+)',string)[0])
        except:
            total_review.append("0")
        try:
            positive_percentage.append(re.findall(r'(\d+)',string)[1])
        except:
            positive_percentage.append("0")
    count += 1
df["total_review"] = total_review
df["positive_percentage"] = positive_percentage
```

```python
import re
years = []
for i in df.release_date:
    string = str(i)
    s = re.findall("(\d{4})",string)[0]
    years.append(s)
df["release_year"] = years
```

```python
df = df.drop(['rating','release_date'],axis=1)
```

#### text cleaning for game_tags

```python
import re
count = 0
for i in df["game_tags"]:
    array = []
    x = re.split("\,",i)
    for k in range(len(x)):
        x[k] = x[k].replace("'", "")
        if k == 0:    
            x[k] = x[k].replace("[", "")
        elif k == len(x)-1: 
            x[k] = x[k].replace("]", "")
            x[k] = x[k].replace(" ", "", 1)
        else:
            x[k] = x[k].replace(" ", "", 1)
        array.append(x[k])
    df["game_tags"][count] = array
    count += 1
```

#### text cleaning for platform_names

```python
import re
count = 0
for i in df["platform_names"]:
    array = []
    x = re.split("\,",i)
    for k in range(len(x)):
        x[k] = x[k].replace("'", "")
        if k == 0:    
            x[k] = x[k].replace("[", "")
        elif k == len(x)-1: 
            x[k] = x[k].replace("]", "")
            x[k] = x[k].replace(" ", "", 1)
        else:
            x[k] = x[k].replace(" ", "", 1)
        x[k] = x[k].replace("platform_img ", "")
        x[k] = x[k].replace("]", "")
        array.append(x[k])
    df["platform_names"][count] = array
    count += 1
```
